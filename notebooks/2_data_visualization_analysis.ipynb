# ðŸ“Š Data Visualization Analysis

Notebook ini menampilkan grafik untuk menyampaikan pola dan tren dari data numerik dan kategorikal.  
Visualisasi sangat penting untuk memahami hubungan antar variabel sebelum mengambil keputusan berbasis data.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/nilai_siswa.csv')
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Nama"], df["Nilai"], color='skyblue')
plt.title("Nilai Siswa")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.grid(True, linestyle="--", alpha=0.3)

# â¬‡ Tambahkan baris ini untuk simpan gambar
plt.savefig('../images/bar_nilai_siswa.png')  # relative path ke folder 'images/'

plt.show()
```

```python
kelas_count = df["Kelas"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(kelas_count, labels=kelas_count.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribusi Kelas Siswa")
plt.axis("equal")

# â¬‡ Tambahkan ini sebelum plt.show()
plt.savefig('../images/pie_kelas_siswa.png')

plt.show()
```
